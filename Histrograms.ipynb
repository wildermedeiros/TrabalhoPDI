```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
baco = cv2.imread('../ni/baco.JPG')
rgbBaco = cv2.cvtColor(baco, cv2.COLOR_BGR2RGB)

ge = cv2.imread('../ni/ge.JPG')
rgbGe = cv2.cvtColor(ge, cv2.COLOR_BGR2RGB)

snow = cv2.imread('../ni/snow.JPG')
rgbSnow = cv2.cvtColor(snow, cv2.COLOR_BGR2RGB)
```

```python
#BGR

histogramValuesB = cv2.calcHist([ge], channels = [0], mask = None, histSize = [256], ranges = [0,256]) 
histogramValuesG = cv2.calcHist([ge], channels = [1], mask = None, histSize = [256], ranges = [0,256]) 
histogramValuesR = cv2.calcHist([ge], channels = [2], mask = None, histSize = [256], ranges = [0,256]) 

plt.plot(histogramValuesB)
plt.plot(histogramValuesG)
plt.plot(histogramValuesR)

```

```python
plt.imshow(rgbGe)
```

```python
plt.xlim([0, 256])
histogram, bin_edges = np.histogram(ge[:, :, 0], bins=256, range=(0, 256))
plt.plot(bin_edges[0:-1], histogram, color="r")
plt.xlabel("Color value")
plt.ylabel("Pixels")
plt.show()

plt.xlim([0, 256])
histogram, bin_edges = np.histogram(ge[:, :, 1], bins=256, range=(0, 256))
plt.plot(bin_edges[0:-1], histogram, color="g")
plt.xlabel("Color value")
plt.ylabel("Pixels")
plt.show()

plt.xlim([0, 256])
histogram, bin_edges = np.histogram(ge[:, :, 2], bins=256, range=(0, 256))
plt.plot(bin_edges[0:-1], histogram, color="b")
plt.xlabel("Color value")
plt.ylabel("Pixels")
plt.show()
```

```python
colors = ("r", "g", "b")
channel_ids = (0, 1, 2)

# create the histogram plot, with three lines, one for
# each color
plt.xlim([0, 256])
for channel_id, c in zip(channel_ids, colors):
    histogram, bin_edges = np.histogram(
        ge[:, :, channel_id], bins=256, range=(0, 256)
    )
    plt.plot(bin_edges[0:-1], histogram, color=c)

plt.xlabel("Color value")
plt.ylabel("Pixels")

plt.show()
```

```python
colors = ("r", "g", "b")
channel_ids = (0, 1, 2)

# create the histogram plot, with three lines, one for
# each color
plt.xlim([0, 256])
for channel_id, c in zip(channel_ids, colors):
    histogram, bin_edges = np.histogram(
        rgbGe[:, :, channel_id], bins=256, range=(0, 256)
    )
    plt.plot(bin_edges[0:-1], histogram, color=c)

plt.xlabel("Color value")
plt.ylabel("Pixels")

plt.show()
```
